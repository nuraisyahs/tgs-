
# Geometrical Meaning of Eigenvalues and Eigenvectors


---

# Introduction

Dalam aljabar linear, **eigenvalues** dan **eigenvectors** menjelaskan bagaimana suatu matriks mentransformasikan vektor.

Transformasi matriks dapat:

- Mengubah panjang vektor
- Mengubah arah vektor

Namun terdapat **arah khusus** yang hanya berubah skalanya.

Arah ini disebut **eigenvector**.

---

# Matrix Transformation

Misalkan matriks

$$
A =
\begin{bmatrix}
0 & 2 \\
1 & 1
\end{bmatrix}
$$

dan vektor

$$
x =
\begin{bmatrix}
1 \\
0
\end{bmatrix}
$$

Transformasi matriks:

$$
y = Ax
$$

---

# Result of Transformation

$$
y =
\begin{bmatrix}
0 & 2 \\
1 & 1
\end{bmatrix}
\begin{bmatrix}
1 \\
0
\end{bmatrix}
=
\begin{bmatrix}
0 \\
1
\end{bmatrix}
$$

Artinya matriks **mengubah posisi dan arah vektor dalam ruang**.

---

# Eigenvalue Equation

Eigenvector memenuhi persamaan:

$$
Av = \lambda v
$$

dimana:

- $v$ = eigenvector  
- $\lambda$ = eigenvalue  

Interpretasi:

- arah vektor tetap
- panjang vektor dikalikan dengan $\lambda$

---

# Eigen Decomposition

Matriks dapat ditulis sebagai

$$
A = V \Lambda V^{-1}
$$

dimana

- **V** = matriks eigenvector
- **Λ** = matriks diagonal eigenvalue

$$
\Lambda =
\begin{bmatrix}
\lambda_1 & 0 \\
0 & \lambda_2
\end{bmatrix}
$$

---

# Geometrical Interpretation

Jika suatu vektor ditulis sebagai kombinasi eigenvector:

$$
x = V\alpha
$$

Maka transformasi menjadi

$$
Ax = V\Lambda\alpha
$$

Artinya:

setiap komponen vektor **dikalikan oleh eigenvalue masing-masing**.

---

# Visual Interpretation

Transformasi matriks pada **unit circle**

$$
x(\theta)=
\begin{bmatrix}
\cos\theta \\
\sin\theta
\end{bmatrix}
$$

Setelah transformasi:

$$
y = Ax
$$

Lingkaran berubah menjadi **ellipse**.

Arah sumbu ellipse adalah **eigenvector**.

---

Interpretasi geometris:

- lingkaran berubah menjadi ellipse
- panjang sumbu ellipse ditentukan oleh **singular values**

---

# Quadratic Function

Pertimbangkan fungsi

$$
f(x) = x^T A x
$$

dengan matriks

$$
A =
\begin{bmatrix}
5 & 3 \\
3 & 5
\end{bmatrix}
$$

---

# Eigenvalues of A

Eigenvalues matriks tersebut

$$
\lambda_1 = 2
$$

$$
\lambda_2 = 8
$$

---

# Visualization Concept

Kontur fungsi

$$
f(x) = x^T A x
$$

akan membentuk **ellipse**.

Eigenvector menunjukkan:

- arah sumbu utama ellipse
- arah perubahan maksimum dan minimum.

---

# Conclusion

Makna geometris eigenvalues dan eigenvectors:

- **Eigenvector** → arah transformasi utama
- **Eigenvalue** → faktor skala pada arah tersebut
- Transformasi matriks mengubah **lingkaran menjadi ellipse**
- Pada fungsi kuadratik, eigenvalues menentukan **nilai ekstrem**
